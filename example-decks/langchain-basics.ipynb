```python
# from langchain_core.output_parsers import StrOutputParser

# Build a chain: prompt -> model -> parse output
# chain = prompt | llm | StrOutputParser()

# result = chain.invoke({
#     "concept": "Retrieval-Augmented Generation",
#     "audience": "beginner"
# })

print("Chain: prompt | llm | output_parser")
print()
print("Result: RAG combines a retrieval system with a language model.")
print("Instead of relying solely on what the model learned during training,")
print("it first searches a knowledge base for relevant documents,")
print("then feeds those documents to the model as context.")
```

## 7. Quick Reference

| Component | Purpose | Example |
|---|---|---|
| **ChatModel** | LLM wrapper | `ChatOpenAI`, `ChatAnthropic` |
| **PromptTemplate** | Reusable prompts | `ChatPromptTemplate.from_messages()` |
| **OutputParser** | Structured output | `StrOutputParser`, `JsonOutputParser` |
| **Chain (LCEL)** | Compose components | `prompt \| llm \| parser` |
| **Memory** | Conversation state | `RunnableWithMessageHistory` |
| **Retriever** | Document search | `vectorstore.as_retriever()` |
| **Agent** | Tool-using LLM | `create_react_agent()` |
